## DSS Black and White images example

Default Value:
-OUTFILE_NAME  SDSS_{0}_{1}_BEST.tiff -MAX_TYPE QUANTILE  -MAX_TYPE QUANTILE  -MAX_LEVEL 0.997 -COLOUR_SAT  7 -MIN_TYPE QUANTILE -MIN_LEVEL 1  -GAMMA_FAC 0.7 

```python
from IPython.core.display import Image
Image("DSS_156.5925946_-34.9634492_BEST.jpeg")
```

Becareful GAMMA spelled with a double M

```python
"i.4".replace(".","_")
```

```python
color_sat = 10
max_level = 0.997
min_level = 1
gama = 0.7
os.system("stiff DSS_2ir_156.5925946_-34.9634492.fits\
            DSS_2r_156.5925946_-34.9634492.fits \
            DSS_2b_156.5925946_-34.9634492.fits -c stiff.conf \
          -OUTFILE_NAME  COLORSAT_{0}_MAX{1}_MIN{3}_GAMAFAC{2}.tiff \
          -COLOUR_SAT  {4} -MAX_TYPE QUANTILE -MAX_LEVEL {5} \
          -MIN_TYPE QUANTILE -MIN_LEVEL {6} -GAMMA_FAC {7}".format(str(color_sat).replace('.','_'),str(max_level).replace('.','_'),str(gama).replace('.','_'),str(min_level).replace('.','_'),color_sat,max_level,min_level,gama))
```

```python
os.system("gm convert  COLORSAT_{0}_MAX{1}_MIN{3}_GAMAFAC{2}.tiff stiff.jpg".format(str(color_sat).replace('.','_'),str(max_level).replace('.','_'),str(gama).replace('.','_'),str(min_level).replace('.','_')))
```

### BEST

Changing Gamma correction factor preserve color , the higher it is the more emphasis on low surface structure. (spread of larger wings of galaxy profile) Conversely, if we want less "fuzziness" in wing then lower correction factor is required.

```python
color_sat = 1
max_level = 1
min_level = 0.99
gama = 0.7
os.system("stiff DSS_2ir_{8}.fits\
            DSS_2r_{8}.fits \
            DSS_2b_{8}.fits -c stiff.conf \
          -COPY_HEADER Y\
          -OUTFILE_NAME  COLORSAT_{0}_MAX{1}_MIN{3}_GAMAFAC{2}.tiff \
          -COLOUR_SAT  {4} -MAX_TYPE QUANTILE -MAX_LEVEL {5} \
          -MIN_TYPE QUANTILE -MIN_LEVEL {6} -GAMMA_FAC {7}".format(str(color_sat).replace('.','_'),str(max_level).replace('.','_'),str(gama).replace('.','_'),str(min_level).replace('.','_'),color_sat,max_level,min_level,gama,coord))
os.system("gm convert  COLORSAT_{0}_MAX{1}_MIN{3}_GAMAFAC{2}.tiff stiff.jpg".format(str(color_sat).replace('.','_'),str(max_level).replace('.','_'),str(gama).replace('.','_'),str(min_level).replace('.','_')))
Image(filename="stiff.jpg")
```
