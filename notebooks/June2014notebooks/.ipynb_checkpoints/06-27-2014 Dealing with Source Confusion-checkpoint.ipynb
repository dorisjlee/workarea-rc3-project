That means we can just ignore the rc3_ra, rc3_dec and just use the ones returned by sextractor.  

```python
import numpy as np
rc3_data = [np.array([0.158083, 28.384556]), np.array([0.195667, 28.402028])] # eliminated because it is too small [0.183333, 28.401444],
#found in test.cat file, and read off from the "radius" list
sextract = [np.array([0.1955781,28.4020029]),np.array([0.1628146,28.3875895])]#eliminated corresponding [0.1835521,28.4014604]

diff = []
coord_match=[]
for i in rc3_data : 
    #determine shift vector 
    for j in sextract:
        print (str(i)+" " +str(j))
        coord_match.append([i,j])
        diff.append(list(j-i))
```
