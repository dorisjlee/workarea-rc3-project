```python
#! /usr/bin/env python
#
# This program identifies pixels that belong to an object of interest, by identifying
# those pixels above some threshold value (i.e., mean + sigma)
#
# Fall 2012 Honor's Project: Big Sky Pictures
#
# Robert J. Brunner and Britt Jeter

import pyfits

#def threshold(image):

    # First you need to get the image mean and the image sigma
    # There is a very easy way, and a longer way.

    #mean = image.mean()
    #sigma = image.std()
    #value = mean + sigma
    
    #ylen, xlen = image.shape
    
    # For this it is easiest to simply make a list of pixels that belong to the object.
    # We need a list for the X pixels and one for the y.
    
    #x_pixels = []
    #y_pixels = []

    
    #for i in range(xlen):
        #for j in range(ylen):
        
            # Now we go through all pixels in the image and determine if they are
            # above the threshold value we just calculated

       #     if (image[j, i] > value):
                
      #          x_pixels.append(i)# If they are we need to keep the x and y indices to mark the objects.
     #           y_pixels.append(j)# Add the x and y coordinates to the appropriate pixel list. (Two lines)

    #return(x_pixels, y_pixels)

from matplotlib import pyplot as pyplt

#def my_plot(x, y):
#    fig = pyplt.figure()
 #   plt = fig.add_subplot(111)
#
 #   # Note I am using Latex in the plot labels
    
  #  plt.scatter(x,y)# Make a scatter plot of x and y. One line 

   # plt.set_xlabel("X")
    #plt.set_ylabel("Y")

    #plt.set_title("ASTR 406: Informatics #2, Problem #1 (Image Pixel Plot)")

    # Uncomment next line to show plot on screen
    # pyplt.show()

    #pyplt.savefig("info2-1-ImageB.pdf")

from matplotlib import pyplot as pyplt

def my_hist_plot(image):

    ylen, xlen = image.shape
    
    # In this method, we trim the pixel list to only hold those pixels
    # less than some maximum value (here mean plus three sigma) in order
    # to focus on the shape of the distribution function near the peak.
    
    mean = image.mean()
    sigma = image.std()
    max_value = mean + 500 * sigma
    
    # This makes a trimmed pixel list by using a list comprehension
    
    pixels = [image[j, i] \
        for i in range(xlen) \
        for j in range(ylen) \
        if image[j, i] < max_value]

    fig = pyplt.figure()
    plt = fig.add_subplot(111)
    
    plt.hist(pixels, bins=1000, log=True)# Plot the histogram (one line)

    plt.set_xlabel("Pixel Value")
    plt.set_ylabel("Pixel Count")

    plt.set_title("Pixel Histogram Plot)")

    # Uncomment next line to show plot on screen
    pyplt.show()

    pyplt.savefig("Coma_i_pixelhist.pdf")
    
# General Main method.

if __name__ == '__main__':

    from sys import argv, exit

    usage = """
    USAGE: python %s image

    WHERE

    image is the name of the fits file containing the image of interest
    (e.g., M86.fits)

    PURPOSE:
	
	This code thresholds an image, and identifies pixels above a threshold value (and thus
	likely part of an object of interest.
    
    """

    # Now if we do not call the function correctly, we should exit and print
    # useful message

    if (len(argv) != 2):
        print usage % (argv[0])
 
    else:
    
        try:
            # Open fits image and get the image pixel data
            fitsImage = pyfits.open(argv[1])
            image = fitsImage[0].data
            
            # Get the pixel lists beloning to the object
            (xp, yp) = threshold(image)
            
            # Plot pixel list
            my_plot(xp, yp)
            
            # Plot Pixel histogram
            my_hist_plot(image)
            
            # Close image
            fitsImage.close()

        except IOError:
            print "ERROR: file ", argv[1], "does not exist.\n"
            exit(-1) # So exit with an error code to the system.    	


            
```
