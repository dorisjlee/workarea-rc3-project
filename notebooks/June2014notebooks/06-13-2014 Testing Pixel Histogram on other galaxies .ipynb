# Orignal pixel histogram 

```python
image = fitsImage[0].data
def my_hist_plot(image):
    image = fitsImage[0].data
    ylen, xlen = image.shape[0],image.shape[1]
    pixels = [image]# Keeping everything
#     print pixels
    fig = pyplt.figure()
    plt = fig.add_subplot(111)
    plt.hist(pixels, bins=1000, log=True)# Plot the histogram (one line)
    plt.set_xlabel("Pixel Value")
    plt.set_ylabel("Pixel Count")
    plt.set_title("Pixel Histogram Plot)")
    pyplt.show()
```

## Emphasis on Low Surface structure

- low minvalue because you want all the light, so no cutting away any low intensity stuff
- lower maxvalue (reasoning?)
- higher gamma value

```python
os.system("stiff SDSS_i_0.645_-3.71055555556.fits  SDSS_r_0.645_-3.71055555556.fits  SDSS_g_0.645_-3.71055555556.fits    -OUTFILE_NAME LOW.tif -c stiff.conf -MAX_TYPE QUANTILE  -MAX_LEVEL 0.98 -COLOUR_SAT  4  -MIN_TYPE QUANTILE -MIN_LEVEL 0  -GAMMA_FAC 1.3 ")
os.system("convert LOW.tif  LOW.jpg")
Image(filename="LOW.jpg")
```
