## Emphasis on Low Surface structure

- low minvalue because you want all the light, so no cutting away any low intensity stuff
- lower maxvalue (reasoning?)
- higher gamma value

```python
os.system("stiff SDSS_i_0.645_-3.71055555556.fits  SDSS_r_0.645_-3.71055555556.fits  SDSS_g_0.645_-3.71055555556.fits    -OUTFILE_NAME LOW.tif -c stiff.conf -MAX_TYPE QUANTILE  -MAX_LEVEL 0.98 -COLOUR_SAT  4  -MIN_TYPE QUANTILE -MIN_LEVEL 0  -GAMMA_FAC 1.5 ")
os.system("convert LOW.tif  LOW.jpg")
Image(filename="LOW.jpg")
```

The speckled colored background comes from too high of a GAMA FACTOR (emphasis low surface structure even for background object) then trying to cut it off with MIN value.

```python
os.system("convert LOW.jpg  LOW.fits")
fitsImage = pyfits.open("LOW.fits")
my_hist_plot(fitsImage) #plot new one here. 
```
