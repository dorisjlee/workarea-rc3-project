Gamma Factor

Power Law : $I=p^\gamma$where I is the true light intensity and p is the pixel value displayed on the computer screen (i.e voltage)

- $\gamma \epsilon$[1.7,2.5], average 2.2
- Remanant of CRT days => for CRT , light intensity and voltage non linear relation 
- LCD follow similar relation
- exponential means more intensity range represented for a smaller range of pixel
- "It is advised to stick to the pure power-law transfer curve for scientific or general use, as STIFF images currently lack the TIFF metadata with the proper transformation tables that would make sRGB and rec."
- gamma correction factor can be used to reduce or enhance low surface brightness without saturating the bright regions 
    - modify $\gamma$ to incr. decr. the contrast scale => affect colors on multichannel images.
    - redefine luminance = $\frac{R+G+B}{3}$ where R,G,B : pixel b4 nonlinear trans
    - then actual r,g,b values (displayed on LCD)
        - $$r \alpha (\frac{R}{Y}Y^{\frac{1}{\gamma Y}})^{\frac{1}{\gamma}}$$
        - vice versa for g,b
        - GAMA_FAC is $\gamma Y$
            - increasing it enhance low surface brightness feature but preserve the color.
